```python
top_left = pd.read_csv('top1-25-1.csv')
top_right = pd.read_csv('top1-25-2.csv')
bottom_left = pd.read_csv('top26-50-1.csv')
bottom_right = pd.read_csv('top26-50-2.csv')
```

```python
bottom_data = pd.concat([bottom_left, bottom_right], axis = 1)
top_data = pd.merge(top_left, top_right, on= ['Track.Name', 'Popularity'])
data = pd.concat([bottom_data, top_data], sort = False)
data

```

```python
data.head()
```

```python
data['NR.'] = range(1,51)

data.set_index("NR.",inplace = True)

data
```

```python
genre = data.groupby("Genre")
```

```python
genre_counted = genre.count()
```

```python
genre_counted[genre_counted['Track.Name'] > 3]
```

```python
most_popular_genre = genre.mean(['Popularity'])
most_popular_genre
```

```python
most_pop = most_popular_genre['Popularity'].max()
most_pop['Genre']
```
